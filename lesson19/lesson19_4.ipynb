```python
import pandas as pd
import numpy as np

p_102= pd.read_csv('各鄉鎮市區人口密度/102.csv')
p_103= pd.read_csv('各鄉鎮市區人口密度/103.csv')
p_104= pd.read_csv('各鄉鎮市區人口密度/104.csv')
p_105= pd.read_csv('各鄉鎮市區人口密度/105.csv')
p_106= pd.read_csv('各鄉鎮市區人口密度/106.csv')
p_107= pd.read_csv('各鄉鎮市區人口密度/107.csv')
p_108= pd.read_csv('各鄉鎮市區人口密度/108.csv')
p_109= pd.read_csv('各鄉鎮市區人口密度/109.csv')
p_110= pd.read_csv('各鄉鎮市區人口密度/110.csv')
p_111= pd.read_csv('各鄉鎮市區人口密度/111.csv')
p_112= pd.read_csv('各鄉鎮市區人口密度/112.csv')

df1=pd.concat([p_102,p_103,p_104,p_105,p_106,p_107,p_108,p_109,p_110,p_111,p_112])

df2 = df1.rename(columns={'statistic_yyy':'統計年',
                    'site_id':'區域別',
                    'people_total':'年底人口數',
                    'area':'土地面積',
                    'population_density':'人口密度'})

df3 = df2.drop(index=0)
f4=df3.dropna()
df5 = df4.set_index('區域別')
def to_int(value):
    try:
        return int(value)
    except:
        return 0

df5['年度人口數'] = df5['年底人口數'].map(to_int)
df5['人口密度'] = df5['人口密度'].map(to_int)
df5
```
