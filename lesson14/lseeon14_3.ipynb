```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
explode = (0.2, 0, 0, 0)

fig, ax = plt.subplots()

ax.pie(values, labels=labels,explode=explode,autopct='%1.3f%%',
       shadow=True,startangle=180,
       colors=['yellow', 'green', 'red', 'blue'])


plt.show()
```
